```python
import ast

env_protest_topic['labels'] = env_protest_topic['labels'].apply(lambda x: ast.literal_eval(x))
env_protest_topic['scores'] = env_protest_topic['scores'].apply(lambda x: ast.literal_eval(x))
```

```python
env_protest_topic["label_scores"] = env_protest_topic.apply(lambda x: dict(zip(x["labels"], x["scores"])), axis=1)
env_protest_topic = pd.json_normalize(env_protest_topic['label_scores'].tolist())
```

```python
env_protest_us = env_protest_us.reset_index(drop=True)
env_protest_topic = env_protest_topic.reset_index(drop=True)
```

```python
env_protest_topic_us = pd.concat([env_protest_us, env_protest_topic], axis = 1)
env_protest_topic_us
```

```python
env_protest_topic_us['topic_env'] = ''
columns = ["mining", "climate", "water", "nuclear", "biodiversity", "oil", "waste"]

for i, row in env_protest_topic_us.iterrows():
    largest_value = 0
    largest_column = ''
    for col in columns:
        if row[col] > largest_value:
            largest_value = row[col]
            largest_column = col
    env_protest_topic_us.at[i, 'topic_env'] = largest_column
```
